# lab 12-3 long sentence 

```python
import tensorflow as tf
import numpy as np
```

```python
sentence = ("if you want to build a ship, don't drum up people together to"
           "collect wood and don't assign them tasks and work, but rather"
           "teach them to long for the endless immensit of the sea.")
```

```python
char_set = list(set(sentence))
char_dic = {c:i for i,c in enumerate(char_set)}
```

```python
data_dim = len(char_set)
hidden_size = len(char_set)
num_classes = len(char_set)
seq_len = 10
learning_rate = 0.1
```

```python
dataX = []
dataY = []
```

```python
for i in range(0,len(sentence)-seq_len):
    x_str = sentence[i:i+seq_len]
    y_str = sentence[i+1:i+seq_len+1]
    print(i, x_str, '->', y_str)
    
    x = [char_dic[c] for c in x_str]
    y = [char_dic[c] for c in y_str]
    
    dataX.append(x)
    dataY.append(y)
```

```python
batch_size = dataX
```

```python
x_one_hot = tf.keras.utils.to_categorical(dataX, num_classes)
y_one_hot = tf.keras.utils.to_categorical(dataY, num_classes)
```

```python
x_one_hot.shape
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.LSTM(units = num_classes, input_shape = (seq_len,num_classes),return_sequences = True))
model.add(tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(units = num_classes, activation = 'softmax')))
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = tf.keras.optimizers.Adam
             (learning_rate = learning_rate),metrics = ['accuracy'])
model.fit(x_one_hot, y_one_hot, epochs = 200)
```

```python
predictions = model.predict(x_one_hot)
```

```python
result_str = []
for i,prediction in enumerate(predictions):
    predict_chr = [char_set[c] for i,c in enumerate(np.argmax(prediction, axis = 1))]
    result_str.append(''.join(predict_chr))
    
```

```python
for i in range(0,len(sentence)-seq_len):
    predict_str = result_str[i]
    y_str = sentence[i+1:i+seq_len+1]
    print("st true -> predict")
    print(i, y_str, '->', predict_str)
```
