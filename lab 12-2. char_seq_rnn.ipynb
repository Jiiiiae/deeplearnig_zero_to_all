# lab 12-2 char_seq_rnn 

```python
import numpy as np
import tensorflow as tf
```

```python
sample = "if you want you"
```

```python
idx2char = list(set(sample))
char2idx = {c:i for i,c in enumerate(idx2char)}

sample_idx = [char2idx[c] for c in sample]
x_data = [sample_idx[:-1]]
y_data = [sample_idx[1:]]
```

```python
dic_size = len(char2idx)
rnn_hidden_size = len(char2idx)
num_classes = len(char2idx)
batch_size = 1
sequence_length = len(sample)-1
learning_rate = 0.1
```

```python
x_one_hot = tf.keras.utils.to_categorical(x_data, num_classes)
y_one_hot = tf.keras.utils.to_categorical(y_data, num_classes)
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.LSTM(units = num_classes, input_shape = x_one_hot.shape[1:],return_sequences = True))
model.add(tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(units = num_classes, activation = 'softmax')))
model.summary()
```

```python
model.input_shape
```

```python
model.compile(loss = 'categorical_crossentropy',optimizer = tf.keras.optimizers.Adam
             (learning_rate = learning_rate),metrics = ['accuracy'])
model.fit(x_one_hot,y_one_hot,epochs = 250)
```

```python
predictions = model.predict(x_one_hot)
```

```python
str=[]
for i,prediction in enumerate(predictions):
    
    result_str = [idx2char[c] for c in np.argmax(prediction,axis=1)]
    
print('\tPrediction: ',''.join(result_str))
```
