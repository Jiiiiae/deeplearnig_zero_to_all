```python
import tensorflow as tf
import numpy as np
```

```python
def min_max_scaler(data):
    numerator = data - np.min(data,0)
    denominator = np.max(data,0) - np.min(data,0)
    
    return numerator / (denominator + 1e-10) #noise term prevents the zero division
```

```python
xy = np.array(
    [
        [828.659973, 833.450012, 908100, 828.349976, 831.659973],
        [823.02002, 828.070007, 1828100, 821.655029, 828.070007],
        [819.929993, 824.400024, 1438100, 818.97998, 824.159973],
        [816, 820.958984, 1008100, 815.48999, 819.23999],
        [819.359985, 823, 1188100, 818.469971, 818.97998],
        [819, 823, 1198100, 816, 820.450012],
        [811.700012, 815.25, 1098100, 809.780029, 813.669983],
        [809.51001, 816.659973, 1398100, 804.539978, 809.559998],
    ]
)
```

```python
xy = min_max_scaler(xy)
x_train = xy[:,0:-1]
y_train = xy[:,-1]
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_dim = x_train.shape[1],activation = 'linear'))
model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(0.1),metrics = ['accuracy'])
model.summary()
```

```python
history = model.fit(x_train,y_train,epochs = 2000)
```

```python
prediction = model.predict(x_train)
evaluation = model.evaluate(x_train,y_train)

print("Prediction: ",prediction)
print("Evaluation: ",evaluation)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
nb_classes = 10

mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test,y_test) = mnist.load_data()

x_train , x_test = x_train/255, x_test/255

y_train, y_test = tf.keras.utils.to_categorical(y_train, nb_classes),tf.keras.utils.to_categorical(y_test, nb_classes)

y_train.shape
```

```python
training_epochs = 15
batch_size = 100
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Flatten(input_shape = (28,28)))
tf.model.add(tf.keras.layers.Dense(10,activation='softmax'))
tf.model.compile(loss='categorical_crossentropy',optimizer=tf.keras.optimizers.SGD(0.1),metrics=['accuracy'])
tf.model.summary()
tf.model.input_shape
```

```python
history = tf.model.fit(x_train,y_train,epochs=5)
```

```python
accuracy = tf.model.evaluate(x_test,y_test)
prediction = tf.model.predict(x_test)
prediction = tf.keras.backend.eval(tf.argmax(prediction,1))
print("Prediction: ", prediction, "Label: ",tf.keras.backend.eval(tf.argmax(y_test,1)))
```

```python
for i in range(10):    
    r = random.randint(0,y_test.shape[0])
    print("Label: ", tf.keras.backend.eval(tf.argmax(y_test[r:r+1,:],1)))
    print("Prediction: ", prediction[r:r+1])
    plt.imshow(x_test[r:r+1,::].reshape(28,28),cmap='Greys')
    plt.show()
```

```python
train_ds = tf.data.Dataset.from_tensor_slices(
    (x_train, y_train)).batch(batch_size)
```

```python
for epoch in range(training_epochs):
    avg_cost = 0
    total_batch = int(x_train.shape[0]/batch_size)
    for it in range(total_batch):
        history = tf.model.fit(train_ds)
        avg_cost += history.hisotory(['loss'])/total_batch
        
    print("Epoch: ", '%04d' % (epoch+1), 'cost =', '{:,9f}.format')
```
